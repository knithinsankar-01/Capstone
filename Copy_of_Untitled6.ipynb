```python
# Step 1: Install required libraries
!pip install transformers pandas

# Step 2: Import necessary libraries
import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Step 3: Prepare the dataset
data = {
    'query': [
        "How can I reset my password?",
        "Where is my order?",
        "How can I update my shipping address?",
        "What is the return policy?",
        "How can I track my order?"
    ],
    'response': [
        "To reset your password, click on 'Forgot Password' on the login page and follow the instructions.",
        "You can check your order status on the 'My Orders' page after logging into your account.",
        "You can update your shipping address in the 'Account Settings' section.",
        "Our return policy allows returns within 30 days of purchase. Please visit our 'Returns' page for more details.",
        "You can track your order using the tracking number provided in the shipping confirmation email."
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Optional: Save dataset to CSV
df.to_csv('customer_support_dataset.csv', index=False)

# Step 4: Load pre-trained GPT-2 model and tokenizer
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Step 5: Define function to generate responses
def generate_response(query):
    input_ids = tokenizer.encode(query, return_tensors='pt')
    output = model.generate(input_ids, max_length=50, num_return_sequences=1, no_repeat_ngram_size=2)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

# Step 6: Implement automated response and personalized engagement
def automated_response(query):
    return generate_response(query)

def personalized_engagement(interaction_history):
    return generate_response(interaction_history)

# Step 7: Testing the functions
query = "How can I reset my password?"
print(f"Customer: {query}\nAI: {automated_response(query)}\n")

customer_interaction_history = "Asked about password reset and order status"
print(f"Customer Interaction: {customer_interaction_history}\nAI (Personalized): {personalized_engagement(customer_interaction_history)}")

# Step 8: Error handling (Optional)
def generate_response_with_error_handling(query):
    try:
        response = generate_response(query)
    except Exception as e:
        response = "Sorry, something went wrong. Please try again later."
        print(f"Error: {e}")
    return response

# Example of error handling
print(generate_response_with_error_handling("How do I update my address?"))
```
